```
import random
```

```
randomNum = random.randint(1,50)

guess = int(input('숫자를 입력하세요: '))
count = 1

while guess != randomNum :
    if guess > randomNum :
        print('더 작은 수를 선택하세요')
    else :
        print('더 큰 수를 선택하세요')
    
    guess = int(input('숫자를 맞춰보세요(1-50): '))
    count += 1
    
print('성공하셨군요', count , '번 만에 맞췄습니다')
print('입력한 숫자는', randomNum, ' 입니다')
```

```
df_data = {
'name': ["Choi", "Choi", "Choi", "Kim", "Park"], 
'year': [2013, 2014, 2015, 2016, 2017], 
'points': [1.5, 1.7, 3.6, 2.4, 2.9]
}                              # 파이썬의 dictionary 로 선언
df2 = pd.DataFrame(df_data) 
df2 
```

```
df2 = pd.DataFrame(df_data, columns=["year", "name", "points", "penalty"],
                                  index=(["one", "two", "three", "four", "five"]))
df2
```

```
df2["penalty"] = 0.5 
df2
```

```
# 연산 후 새로운 열을 추가하기
df2["net_points"] = df2["points"] - df2["penalty"]
df2
```

```
# 조건 연산으로 열 추가
df2["high_points"] = df2["net_points"] > 2.0 
df2
```

```
df2['zeros'] = np.arange(5) 
df2
```

```
# 열 삭제
del df2["zeros"] 
df2
```

```
plt.title('x축, y축 범위 설정')
plt.plot([10,20,30,40],[1,4,9,16], 
        c='b', lw=5, ls='--', marker='o',ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0,50)
plt.ylim(-10,30)
plt.grid(True)     # 그리드 나타내기
plt.show()
```

```
import numpy as np

t = np.arange(0., 5., 0.5)
plt.title('라인 플롯에서 여러개 선 그리기')
plt.plot(t, t, 'r--', t, 0.5* t**2, 'bs:', t, 0.2* t**3, 'g^-')
plt.show()
```

```
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)   # upper left
plt.show()
```

```
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```
